```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ecom.csv')
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
#Target Variable Churn

print("\n Churn Value Counts:\n" , df['Churn'].value_counts(normalize=True))
sns.countplot(x = 'Churn'  , data = df)
plt.title('Churn Value Counts')
plt.show()
```

```python
#Tenure Distribution
sns.histplot(df['Tenure'] , kde=True)
plt.xlabel('Tenure')
plt.ylabel('Frequency')
plt.title('Tenure Distribution')
plt.show()
```

```python
#cityTier distribution
sns.countplot(x = 'CityTier'  , data = df)
plt.title('cityTier Distribution')
plt.show()
```

```python
#satisfaction distribution
sns.countplot(x = 'SatisfactionScore'  , data = df)
plt.title('Satisfaction Distribution')
plt.show()
```

```python
#gender distribution
print("\n Gender Distribution:\n" , df['Gender'].value_counts(normalize=True))
sns.countplot(x = 'Gender'  , data = df)
plt.title('Gender Value Counts')
plt.show()
```

```python
#Churn vs Tenure
sns.boxplot(x = 'Churn' , y = 'Tenure' , data = df)
plt.title('Churn vs Tenure')
plt.show()
```

```python
#Churn vs CityTier
sns.countplot(x = 'Churn' , hue = 'CityTier' , data = df)
plt.title('Churn vs CityTier')
plt.show()
```

```python
#Churn vs SatisfactionScore
sns.countplot(x = 'Churn' , hue = 'SatisfactionScore' , data = df)
plt.title('Churn vs SatisfactionScore')
plt.show()
```

```python
#churn vs PreferredLoginDevice
sns.countplot(x = 'Churn' , hue = 'PreferredLoginDevice' , data = df)
plt.title('Churn vs PreferredLoginDevice')
plt.show()
```

```python
#Churn vs PerferredPaymentMode
sns.countplot(x = 'Churn' , hue = 'PreferredPaymentMode' , data = df)
plt.title('Churn vs PreferredPaymentMode')
plt.show()
```
